### 5. Stacking and Concatenation
Interviews often test how you combine datasets. `np.concatenate`, `np.vstack` (vertical), and `np.hstack` (horizontal) are key.

```python
a=np.array([1,2,3])
b=np.array([4,5,6])                 #using concatenation
a,b
```

```python
combine=np.concatenate([a,b])
combine
```

```python
a = np.array([[1, 2, 3],["P","q","r"]])
b = np.array([[4, 5, 6],[9,0,8]])

print("Horizontal Stack:", np.hstack((a, b)))
print("Vertical Stack:\n", np.vstack((a, b)))
print(np.stack((a,b),axis=1))
```
