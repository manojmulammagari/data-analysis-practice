```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
gap=px.data.gapminder()
gap
```

```python
temp=gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
#  heat map is a  axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp)
```

```python
temp=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(10,10))
sns.heatmap(temp,annot=True,cmap='viridis')
```

#Boxplot

### What is a Box Plot?
A box plot (also known as a box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.

### Components of a Box Plot:
*   **Median (Q2/50th Percentile)**: The middle value of the dataset, dividing it into two equal halves. This is represented by the line inside the box.
*   **First Quartile (Q1/25th Percentile)**: The middle value of the first half of the dataset. The bottom of the box indicates Q1.
*   **Third Quartile (Q3/75th Percentile)**: The middle value of the second half of the dataset. The top of the box indicates Q3.
*   **Interquartile Range (IQR)**: The range between the first and third quartiles (IQR = Q3 - Q1). This represents the middle 50% of the data.
*   **Whiskers**: These extend from the box to the minimum and maximum values within a certain range, typically 1.5 times the IQR from Q1 and Q3. Values outside this range are considered outliers.
*   **Outliers**: Data points that fall outside the whiskers, often plotted as individual points (e.g., circles or asterisks). These can indicate unusual or extreme values in the dataset.

### What Box Plots are Used For:
*   **Understanding Data Distribution**: Quickly see the spread and central tendency of a dataset.
*   **Identifying Skewness**: Observe if the data is symmetric or skewed (e.g., if the median is closer to one end of the box or if one whisker is much longer).
*   **Detecting Outliers**: Easily spot data points that are significantly different from the rest of the dataset.
*   **Comparing Distributions**: Compare the distributions of multiple datasets side-by-side (e.g., sales across different regions) to understand differences in their central tendency, spread, and presence of outliers.


```python
sns.boxplot(y='total_bill',data=tips)
```

```python
sns.boxplot(x='sex',y='total_bill',data=tips)
```

```python
# Catplot -> figure level function
sns.catplot(x='day',y='total_bill',data=tips,kind='box')
```

```python
# hue
sns.catplot(x='day',y='total_bill',data=tips,kind='box',hue='sex')
```
