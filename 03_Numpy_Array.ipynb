#Broadcasting
it'll perform if there is a scalar shape or a common dimension in both arrays

```python
image=np.array([[0,2,3],[4,5,6],[7,8,9]])
image
```

```python
a=np.arange(3).reshape(1,3)
b=np.arange(3)
print(a)
print(b)
print(a+b)
```

```python
a=np.arange(3).reshape(3,1)
b=np.arange(3)
print(a)
print(b)
print(a+b)
```

When attempting to perform `a + b` with `a.shape = (1, 4)` and `b.shape = (3,)`, it will result in an error due to NumPy's broadcasting rules.

Here's why:

1.  **Dimension Alignment**: NumPy compares array shapes element-wise, starting from the trailing (rightmost) dimension.
    *   `a`: `(1, 4)`
    *   `b`: `(  3)` (internally, NumPy treats it as `(1, 3)` for comparison by padding with ones on the left if dimensions differ).

2.  **Compatibility Check**: For broadcasting to work, dimensions must either be equal, or one of them must be 1.
    *   **Rightmost dimension**: `4` (from `a`) vs `3` (from `b`). These are neither equal nor does one of them equal 1. Therefore, broadcasting fails.

To make these arrays broadcastable, you need to ensure their dimensions are compatible according to the rules. For instance, if you change `b` to have a shape of `(3, 1)`:
*   `a`: `(1, 4)`
*   `b`: `(3, 1)`

Now, comparing from right to left:
*   Rightmost dimension: `4` (from `a`) vs `1` (from `b`). `b`'s dimension `1` will stretch to `4`.
*   Next dimension: `1` (from `a`) vs `3` (from `b`). `a`'s dimension `1` will stretch to `3`.

The resulting broadcasted shape would be `(3, 4)`, and the operation would succeed.

```python
a=np.arange(4).reshape(1,4)
b=np.arange(3).reshape(3)
print(a)
print(b)
# print(a+b)
```

### Vectorization vs Loops
Vectorization allows you to perform batch operations without explicit `for` loops.

```python
import time

# Large array for timing
large_arr = np.arange(1000000)

# Using a loop
start = time.time()
loop_result = [x + 5 for x in large_arr]
print(f"Loop time: {time.time() - start:.4f}s")

# Using vectorization
start = time.time()
vec_result = large_arr + 5
print(f"Vectorized time: {time.time() - start:.4f}s")
```

### Cosine Similarity Implementation
We can calculate cosine similarity using NumPy's dot product and linear algebra norm functions.

```python
def cosine_similarity(v1, v2):
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    return dot_product / (norm_v1 * norm_v2)

# Example vectors
vec_a = np.array([1, 2, 3])
vec_b = np.array([1, 2, 3])
vec_c = np.array([-1, -2, -3])

print(f"Similarity (same): {cosine_similarity(vec_a, vec_b):.2f}")
print(f"Similarity (opposite): {cosine_similarity(vec_a, vec_c):.2f}")
```
