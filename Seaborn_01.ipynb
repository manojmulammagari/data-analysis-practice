#Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
#Scatter plot -> Axes level function
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
#relplot -> figure level -> square shape
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='smoker',col='time',kind='scatter',size="size") # kind is the plot we want
```

```python
gap = px.data.gapminder()
temp=gap[gap['country']=='India']
temp
```

```python
sns.lineplot(data=temp,x='year',y='lifeExp')
```

```python
sns.relplot(data=temp,x='year',y='lifeExp',kind='line')
```

```python
temp_df=gap['country'].isin(['India','United States','Pakistan'])
temp_df=gap[temp_df]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country')
```

```python
# Facet plot -> only work with figure level function (Relplot)
#  it cant work with lineplot or scatterplot
sns.relplot(kind='scatter',data=tips,x='total_bill',y='tip',col='sex',row='smoker') # i can use row instead of col too
```

# Distributed plots
figlevel functions -> displot

axes level -> histplot -> kdeplot -> rugplot

```python
#plotting univariate histgram
tips
```

```python
sns.histplot(data=tips,x='total_bill',kde=True)
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
sns.histplot(data=tips,x='day',kde=True)
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='hist')
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic,x='age',kind='hist',hue='sex',multiple='stack')# i can use elements='step' instead of stack
```

```python
sns.displot(data=titanic,x='age',kind='hist',col='sex',multiple='stack',kde=True)
```
