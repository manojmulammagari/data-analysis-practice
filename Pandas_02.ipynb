### Renaming Columns and Converting to Datetime

To rename specific columns, you can use a dictionary with the mapping `{old_name: new_name}`.

To convert a string or numeric column to actual datetime objects, use `pd.to_datetime()`.

```python
# 1. Renaming columns
df = df.rename(columns={'sample': 'grade', 'y_income_thousands_usd': 'marks'})

# 2. Converting to datetime
# Note: If you had a 'date' column string like '2023-01-01'
# df['date_column'] = pd.to_datetime(df['date_column'])

display(df.head())
```

```python
import pandas as pd
import numpy as np

# Create a small example with a NaN
demo_df = pd.DataFrame({
    'A': [10, 20, np.nan],
    'B': [5, 5, 5]
})

print("1. Original DataFrame:")
display(demo_df)

print("\n2. Sum with axis=0 (Calculates sum for each COLUMN):")
# This collapses rows, giving a result for each column
display(demo_df.sum(axis=0))

print("\n3. Sum with axis=1 (Calculates sum for each ROW):")
# This collapses columns, giving a result for each row
display(demo_df.sum(axis=1))

print("\nNote: If you use axis=0 on the main 'df', the 'grade' column (strings) might cause issues in some versions or return NaN/empty results for math.")
```

```python
# Example of using axis=0
# We will calculate the average (mean) for each numeric column

# We select only the numeric columns first to avoid errors with strings
numeric_df = df[['x_education_years', 'marks']]

print("Average values for each COLUMN (using axis=0):")
column_averages = numeric_df.mean(axis=0)
display(column_averages)

print("\nMaximum values for each COLUMN (using axis=0):")
column_max = numeric_df.max(axis=0)
display(column_max)
```

```python
print(f"Type of df['marks']: {type(df['marks'])}")
print(f"Type of df: {type(df)}")

# This works because it's a Series (1D)
test_series = df['marks'].apply(lambda x: x * 2)
print("\nSeries apply worked without axis.")

try:
    # This would fail because the DataFrame (2D) doesn't know which direction to go
    df.apply(lambda x: x * 2)
except Exception as e:
    print(f"\nDataFrame apply failed without axis: {e}")
```

```python
def pass_fail(score):
  if score>=20:
    return "Pass"
  else:
    return "Fail"

df["Results"]=df["marks"].apply(pass_fail)
df.head()
```

```python
import random

# 1. We define the list of possible values
grades = ['A', 'B', 'C', 'D', 'E']

# 2. We use .apply() with a lambda function
# - 'lambda row:' is a small function that runs for every row
# - 'random.choice(grades)' picks one random letter from our list
# - 'axis=1' tells pandas to move through the table row by row
df['grade'] = df.apply(lambda x: random.choice(grades), axis=1)

# 3. Show the results
display(df.head())
```

### Using .apply(axis=1) across multiple columns

When we use `axis=1`, the variable `row` in our function represents a single horizontal record from the table.

```python
# Example 1: Using a Lambda to combine columns
# Let's create a 'Status' string using both 'grade' and 'Results' columns
# df['Summary'] = df.apply(lambda r: f"Grade {r['grade']} was a {r['Results']}", axis=1)

# Example 2: Using a Defined Function for complex multi-column logic
def calculate_bonus(r):
    # Logic using two different columns
    if r['marks'] > 40 and r['x_education_years'] > 15:
        return "High Performance"
    else:
        return "Standard"

df['Performance_Tag'] = df.apply(calculate_bonus, axis=1)

display(df.head())
```

```python
grade_mapping={
    'A':"Excellent",
    'B':"Good",
    'C':"Average",
    'D':"Below Average",
    'E':"Fail/Need to practice more"
}

df["grade_description"]=df["grade"].map(grade_mapping)
df[["obs","grade",'Results','grade_description']].head(7)

```
