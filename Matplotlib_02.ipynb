```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np

dept_list = ['Sales', 'Marketing', 'Engineering', 'HR', 'Finance']
salaries = np.random.randint(40000, 120000, 20)
dept = np.random.choice(dept_list, 20)

df = pd.DataFrame({
    'Department': dept,
    'Salary': salaries
})

df.head()
```

```python
df.info()
```

```python
count=df["Department"].value_counts()
count
```

```python
plt.pie(count, labels=count.index, autopct="%1.0f%%",explode=[0.2,0.6,0,0,0])
plt.axis("equal")
plt.show()
```

```python
# Countplot
plt.bar(count.index,count,color=["RED","Black","green","pink"])
plt.show()
```

```python
# Bivariate - numerical -numerical
```

```python
# Sorting salaries and assigning ages in a increasing order to show correlation
df['Salary'] = np.sort(df['Salary'])
df['Age'] = np.sort([10,32,55,43,66,55,49,45,23,77,56,36,29,16,59,72,64,48,61,27])
display(df.head())
```

```python
# scatter plot
plt.scatter(df["Age"],df['Salary'],color="orange")
plt.show()
```

```python
# line plot
plt.plot(df["Age"],df['Salary'],color="red")
plt.show()
```

```python
plt.bar(df["Age"],df["Salary"],color="Red",align="center")
plt.show()
```

```python
#  Bivariate - numerical-categorical
```

```python
import matplotlib.pyplot as plt

data_to_plot = [df[df['Department'] == d]['Salary'] for d in df['Department'].unique()]
labels = df['Department'].unique()

plt.figure(figsize=(10, 6))
plt.boxplot(data_to_plot, tick_labels=labels, patch_artist=True)
plt.title('Salary Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Salary')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
#  pie chart :
s_b_d=df.groupby("Department")["Salary"].sum()
s_b_d
```

```python
plt.pie(s_b_d,labels=s_b_d.index,autopct="%1.0f%%")
plt.show()
```

```python
mean_salaries = df.groupby('Department')['Salary'].mean()

plt.figure(figsize=(10, 6))
plt.bar(mean_salaries.index, mean_salaries, color=['skyblue', 'orange', 'lightgreen', 'salmon', 'lightblue'])
plt.title('Mean Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.grid()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
