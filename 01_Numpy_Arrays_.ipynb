```python
import numpy as np
```

```python
# Creation: arange generates values within a given interval
# Syntax: np.arange(start, stop, step)
a = np.arange(1, 21, 2)
display(a)
```

```python
# Vectorization: Applying operations to the whole array without explicit loops
b = a**2  # Calculates the square of every element
display(b)
```

```python
# linspace: Generates 'num' evenly spaced samples over a closed interval [start, stop]
arr = np.linspace(1, 10, 4)
display(arr)
```

```python
# logspace: Returns numbers spaced evenly on a log scale
# Below creates 3 numbers between 10^1 and 10^2
np.logspace(1, 2, 3)
```

```python
# Pre-allocation: Creating arrays filled with zeros
# Useful for initializing weights or placeholders
np.zeros([2, 3])
```

```python
## 2. Pre-allocated Arrays
# Create an array of shape [2, 2] filled with 1.0 (float by default)
np.ones([2, 2])
```

```python
# Specifying dtype: Useful for saving memory when precision isn't required
np.ones([2, 2, 4], dtype=int)
```

```python
# np.full: Useful for initializing a bias vector or constant filters
np.full(10, 3)
```

```python
# Create a 3x3 matrix filled with the number 6
arr = np.full([3, 3], 6)
display(arr)
```

```python
# np.empty: Allocates memory without initializing, making it the fastest option for pre-allocation
b = np.empty([2, 3])
display(b)
```

```python
## 3. Random Sampling
# rand: Generates 10 values from a uniform distribution between [0, 1)
np.random.rand(10)
```

```python
# 2x3 matrix of random floats from a uniform distribution
np.random.rand(2, 3)
```

```python
# randn: Samples from the Standard Normal Distribution (mean=0, std=1)
np.random.randn(3, 3)
```

```python
# Random Integers: Generates random integers from low (inclusive) to high (exclusive)
# Syntax: np.random.randint(low, high, size)
np.random.randint(1, 100, 7)
```

```python
# randint: Random integers in range [1, 100) with a 3x4 shape
np.random.randint(1, 100, (3, 4))
```

```python
## 4. Data Types and Casting
# Note: NumPy arrays must have a single data type (homogeneity)
# Combining ints and strings results in 'Unicode' type (<U21)
my_list = [1, 2, 3, "MAnoj"]
arr = np.array(my_list)
display(arr)
```

```python
## 4. Data Types and Casting
# Explicit dtype conversion during array creation
a = np.array([1, 2, 3.1, 3, "5"], dtype=str)
display(a)
```

```python
arr=np.array([1,2,3,4,5])
arr
```

```python
# astype: Used for explicit type conversion
# Converting integer array to float for scientific calculations
arr_numeric = np.array([1, 2, 3, 4, 5])
new_arr = arr_numeric.astype(np.float64)
display(new_arr)
```

```python
# String conversion: Useful for data cleaning and export preparations
arr2 = new_arr.astype(np.str_)
display(arr2)
```

```python
# Type casting limitations
# Attempting to convert 'Hello' to an integer would raise a ValueError
a = np.array(["1", "2", "Hello"])
display(a)
# a.astype(np.int64)  # This line is commented out because it would fail
```

```python
## 2. Array Attributes
# Understanding the metadata of an array object
new = np.array([[1, 2, 3], [4, 5, 6]])
display(new)

print(f"Dimensions (ndim): {new.ndim}")
print(f"Shape (rows, cols): {new.shape}")
print(f"Total elements (size): {new.size}")
print(f"Total memory (bytes): {new.nbytes}")
```

```python
# ndim: Number of axes (dimensions) of the array
print(f"Number of dimensions: {new.ndim}")
```

```python
# shape: A tuple of integers indicating the size of the array in each dimension
print(f"Array shape: {new.shape}")
```

```python
# size: Total number of elements in the array (rows * cols)
print(f"Total size: {new.size}")
```

```python
# Calculating total memory consumption manually
# total_bytes = itemsize * size
total_memory = new.itemsize * new.size
print(f"Manual memory calculation: {total_memory} bytes")
```
