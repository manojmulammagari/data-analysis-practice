
Numpy Data Types and Type Casting

```python
list=[1,2,3,"MAnoj"]
arr=np.array(list)
arr
```

```python
a=np.array([1,2,3.1,3,"5"],dtype=str)       #dtype converts the list into given datatype
a
```

```python
arr=np.array([1,2,3,4,5])
arr
```

```python
new_arr=arr.astype(np.float64)
new_arr
```

```python
arr2=new_arr.astype(np.str_)
arr2
```

```python
# Type casting errors
a=np.array(["1","2",'Hello'])
a
# a.astype(np.int64) it'll give an error it can't convert hello into integer
```
